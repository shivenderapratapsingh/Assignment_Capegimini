```python
import pandas as pd
import numpy as np

customers_df = pd.read_csv("Customers1 - Sheet1 (1).csv")
sales_df     = pd.read_csv("Sales1 - Sheet1 (1).csv")
support_df   = pd.read_csv("support1 - Sheet1 (1).csv")

```

```python
#CLEAN COLUMN NAMES (BEST PRACTICE)

for df in [customers_df, sales_df, support_df]:
    df.columns = df.columns.str.strip()
    df.columns = df.columns.str.replace(" ", "_")
    df.columns = df.columns.str.lower()
```

```python
# 3Ô∏è‚É£ HANDLE MISSING VALUES
# ==============================
customers_df["age"] = customers_df["age"].fillna(customers_df["age"].median())
```

```python
# 4Ô∏è‚É£ NUMPY BROADCASTING
# ==============================
sales_df["discounted_price"] = sales_df["price"].values * 0.90
sales_df["revenue"] = sales_df["quantity"] * sales_df["price"]
```

```python
# 5Ô∏è‚É£ DATE CONVERSION
# ==============================
sales_df["orderdate"] = pd.to_datetime(sales_df["orderdate"])
customers_df["signupdate"] = pd.to_datetime(customers_df["signupdate"])
```

```python
# January 2025 orders
jan_2025_orders = sales_df[sales_df["orderdate"].dt.to_period("M") == "2025-01"]
```

```python
# 6Ô∏è‚É£ FILTERING
# ==============================
north_customers = customers_df[customers_df["region"] == "North"]
high_revenue_orders = sales_df[sales_df["revenue"] > 10000]
```

```python
# 7Ô∏è‚É£ SORTING
# ==============================
customers_sorted = customers_df.sort_values("signupdate")
sales_sorted = sales_df.sort_values("revenue", ascending=False)
```

```python
# 8Ô∏è‚É£ GROUPING
# ==============================
sales_with_region = sales_df.merge(
    customers_df[["customerid", "region"]],
    on="customerid",
    how="left"
)

avg_region_rev = sales_with_region.groupby("region")["revenue"].mean().reset_index()
avg_res_time = support_df.groupby("issuetype")["resolutiontime"].mean().reset_index()
```

```python
# 9Ô∏è‚É£ MERGING ALL DATA
# ==============================
full_df = customers_df.merge(sales_df, on="customerid", how="left")
full_df = full_df.merge(support_df, on="customerid", how="left")
```

```python
# üîü FEATURE ENGINEERING
# ==============================
clv = sales_df.groupby("customerid")["revenue"].sum().reset_index()
clv.rename(columns={"revenue": "clv"}, inplace=True)
avg_res_customer = support_df.groupby("customerid")["resolutiontime"].mean().reset_index()
avg_res_customer.rename(columns={"resolutiontime": "avg_resolution_time"}, inplace=True)

full_df = full_df.merge(clv, on="customerid", how="left")
full_df = full_df.merge(avg_res_customer, on="customerid", how="left")
```

```python
# 1Ô∏è‚É£1Ô∏è‚É£ EXPORT
# ==============================
full_df.to_csv("Cleaned_Data.csv", index=False)

print("‚úÖ Cleaned_Data.csv created successfully.")
```
