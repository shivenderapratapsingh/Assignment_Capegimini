```python
import numpy as np
from sklearn.neural_network import MLPClassifier

# Training data (Features)
# [Videos Watched, Time Spent]
X = np.array([
    [2, 15],
    [3, 20],
    [8, 60],
    [9, 75]
])

# Target labels
# 0 = Dropout
# 1 = Complete Course
y = np.array([0, 0, 1, 1])

# Create Neural Network Model
model = MLPClassifier(hidden_layer_sizes=(4,), 
                      activation='relu', 
                      solver='adam', 
                      max_iter=500)

# Train the model
model.fit(X, y)

# New student data
new_student = np.array([[7, 55]])

# Prediction
prediction = model.predict(new_student)

print("Prediction:", prediction)

if prediction[0] == 1:
    print("Student will complete the course")
else:
    print("Student will drop out")
```
