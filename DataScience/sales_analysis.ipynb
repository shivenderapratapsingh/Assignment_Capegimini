```python
import pandas as pd
```

```python
# load dataset
sales_df = pd.read_csv("sample_data1 - Sheet1 (1).csv")
```

```python
# Basic overview
sales_df.info()
sales_df.head()
```

```python
# Missing values check
sales_df.isna().sum()
```

```python
# Remove rows with missing Sales
sales_clean = sales_df.dropna(subset=['Monthly_Sales'])
```

```python
# Total sales by Region
sales_clean.groupby('Region')['Monthly_Sales'].sum().sort_values(ascending=False)
```

```python
# Average sales by Region
sales_clean.groupby('Region')['Monthly_Sales'].mean()
```

```python
# Transaction count per Region
sales_clean['Region'].value_counts()
```
