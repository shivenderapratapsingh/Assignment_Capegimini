```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("employee_attrition - Sheet1.csv")
```

```python
# Attrition Rate
total = df.groupby("department")["attrition"].count()
left = df[df["attrition"] == "Yes"].groupby("department")["attrition"].count()

attrition_rate = (left / total) * 100

print(attrition_rate.sort_values(ascending=False))
```

```python
#Average Salary of Employees Who Left
avg_salary_left = df[df["attrition"] == "Yes"]["monthly_income"].mean()
print("Average Salary (Employees who left):", avg_salary_left)
```

```python
#Salary Comparison (Left vs Stayed)
salary_comparison = df.groupby("attrition")["monthly_income"].mean()
print(salary_comparison)
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(" ", "")
df.columns = df.columns.str.lower()

print(df.columns)
```

```python
#Years at Company vs Attrition
years_analysis = df.groupby("attrition")["years_at_company"].mean()
print(years_analysis)
```

```python
#Job Satisfaction vs Attrition
job_analysis = df.groupby("attrition")["job_satisfaction"].mean()
print(job_analysis)
```

```python
#Plot Attrition by Department
plt.figure()
attrition_rate.plot(kind="bar")
plt.title("Attrition Rate by Department")
plt.ylabel("Attrition Rate (%)")
plt.xlabel("Department")
plt.xticks(rotation=45)
plt.show()
```
