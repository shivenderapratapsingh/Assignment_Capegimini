```python
from sklearn.tree import DecisionTreeClassifier,export_text
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("University Dataset - Sheet1.csv")
# Step 2: Define Features and Target
X = df[['HighSchool_GPA', 'Exam_Score', 'Extracurriculars']]
y = df['Admission_Label']
```

```python
# Step 3: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Step 4: Train Decision Tree
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Step 5: Predictions & Accuracy
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)

# Step 6: Display Rulebook
feature_names = ['HighSchool_GPA', 'Exam_Score', 'Extracurriculars']
tree_rules = export_text(model, feature_names=feature_names)
print("\nUniversity Admission Rulebook:\n")
print(tree_rules)

# Step 7: Test New Applicant
# Example: GPA=3.5, Exam=1250, Extracurriculars=1
new_student = [[3.5, 1250, 1]]
prediction = model.predict(new_student)
print("\nNew Student Prediction (1=Accepted, 0=Rejected):", prediction[0])
```
