```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate 1000 salaries
salaries = np.random.normal(loc=50000, scale=10000, size=1000)

# Plot histogram
plt.figure()
plt.hist(salaries, bins=30)
plt.title("Salary Distribution of Employees")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.show()
```
