```python
# Scenario Question

# A biomedical researcher is designing a convolutional neural network (CNN) to analyze microscopic grayscale 
# images of blood cells (64×64 pixels). She begins with a single convolutional layer defined as follows:

# Input: 64×64 grayscale images (1 channel)
# Convolutional Layer: 16 filters, each of size 5×5
# Stride: 1
# Padding: Same
# Activation Function: ReLU
# The goal of this layer is to extract local structural features such as cell boundaries and texture variations, which will later help in classifying whether the blood cells are healthy or show signs of abnormality.
```

```python
import torch
import torch.nn as nn

# Single Conv2D layer
conv = nn.Conv2d(
    in_channels  = 1,    # grayscale input
    out_channels = 16,   # 16 filters -> 16 feature maps
    kernel_size  = 5,    # 5x5 filter
    stride       = 1,
    padding      = 2     # SAME padding: output = same size
)

# Activation function
relu = nn.ReLU()

# Forward pass
x = torch.randn(1, 1, 64, 64)   # (batch, C, H, W)

out = conv(x)
out = relu(out)

print(out.shape)   # torch.Size([1, 16, 64, 64])

# Inspect learnable parameters
print('Weights:', conv.weight.shape)   # (16, 1, 5, 5)
print('Bias   :', conv.bias.shape)     # (16,)

# Total parameters
print('Total  :', 16*1*5*5 + 16)       # = 416
```

```python
import tensorflow as tf
from tensorflow import keras

# Single Conv2D layer
conv = keras.layers.Conv2D(
    filters     = 16,          # 16 filters -> 16 feature maps
    kernel_size = (5, 5),      # 5x5 filter
    strides     = (1, 1),
    padding     = 'same',      # SAME padding
    activation  = 'relu',
    input_shape = (64, 64, 1)  # 64x64 grayscale image
)

# Build and summarise
model = keras.Sequential([conv])
model.build(input_shape=(None, 64, 64, 1))
model.summary()

# Output shape: (None, 64, 64, 16)
# Trainable params: 416
```
