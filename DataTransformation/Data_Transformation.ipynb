```python
import pandas as pd
df = pd.read_csv("Dataset  - Sheet1 (1).csv")
```

```python
df.columns = df.columns.str.strip()

print("Columns:", df.columns)
print("\nFirst 5 rows:")
print(df.head())
```

```python
# 1️⃣ Exceptional performers (Score > 90)
exceptional = df[df["Score"] > 90]
print("\nEmployees with Score > 90:")
print(exceptional)
```

```python
# 1️⃣ Exceptional performers (Score > 90)
exceptional = df[df["Score"] > 90]
print("\nEmployees with Score > 90:")
print(exceptional)
```

```python
# 2️⃣ Records for Sales department only
sales_dept = df[df["Department"] == "Sales"]
print("\nSales Department Records:")
print(sales_dept)
```

```python
# 3️⃣ Sort by Score (Descending)
sorted_score = df.sort_values(by="Score", ascending=False)
print("\nEmployees Ranked by Score (High to Low):")
print(sorted_score)
```

```python
# 4️⃣ Sort by Department, then Score
dept_sorted = df.sort_values(by=["Department", "Score"], ascending=[True, False])
print("\nDepartment-wise Ranking:")
print(dept_sorted)
```

```python
# 5️⃣ Average Score per Department
dept_avg = df.groupby("Department")["Score"].mean()
print("\nAverage Score per Department:")
print(dept_avg)
```

```python
# 6️⃣ Maximum Score per Quarter
quarter_max = df.groupby("Quarter")["Score"].max()
print("\nMaximum Score per Quarter:")
print(quarter_max)
```
