```python
# Agent 1: Crisis Coordinator (Broadcaster)
# - Broadcasts: “Data breach detected in customer database. Immediate response required.”
# - Sends this to all other agents simultaneously.

# 🛡️ Agent 2: IT Security Agent
# - Receives broadcast.
# - Responds: “Isolate affected servers, patch vulnerabilities, start forensic analysis.”

# 📞 Agent 3: Communications Agent
# - Receives broadcast.
# - Responds: “Draft internal memo, prepare press release, notify stakeholders.”

# 💰 Agent 4: Finance Agent
# - Receives broadcast.
# - Responds: “Estimate financial impact, allocate emergency funds, review insurance coverage.”

# 👩‍⚖️ Agent 5: Legal Agent
# - Receives broadcast.
# - Responds: “Assess regulatory obligations, prepare compliance reports, advise on liability.”

# 👩‍💼 Agent 6: HR Agent
# - Receives broadcast.
# - Responds: “Brief employees, provide guidance on handling customer queries, ensure morale support.”

# 🧑‍⚖️ Agent 7: Decision Agent (Coordinator)
# - Collects all responses.
# - Integrates into a final crisis response plan:
# “Servers isolated, communications prepared, financial impact assessed, compliance secured, employees briefed.”
```

```python
import os
import requests
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("NVIDIA_API_KEY")

def call_llm(prompt):
    url = "https://integrate.api.nvidia.com/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    data = {
        "model": "meta/llama3-70b-instruct",
        "messages": [{"role": "user", "content": prompt}],
        "temperature": 0.3
    }

    res = requests.post(url, headers=headers, json=data)
    return res.json()["choices"][0]["message"]["content"]
```

```python
def market_research_worker(goal):
    prompt = f"""
    You are a Market Research Expert.

    Goal: {goal}

    Analyze:
    - Demand
    - Competition
    - Customer preferences

    Give short insights.
    """
    return call_llm(prompt)


def finance_worker(goal):
    prompt = f"""
    You are a Finance Expert.

    Goal: {goal}

    Analyze:
    - Budget required
    - ROI
    - Pricing strategy

    Give short insights.
    """
    return call_llm(prompt)


def operations_worker(goal):
    prompt = f"""
    You are an Operations Expert.

    Goal: {goal}

    Analyze:
    - Supply chain
    - Production capacity
    - Logistics

    Give short insights.
    """
    return call_llm(prompt)


def legal_worker(goal):
    prompt = f"""
    You are a Legal Expert.

    Goal: {goal}

    Analyze:
    - Regulations
    - Compliance
    - IP issues

    Give short insights.
    """
    return call_llm(prompt)


def hr_worker(goal):
    prompt = f"""
    You are an HR Expert.

    Goal: {goal}

    Analyze:
    - Hiring needs
    - Training
    - Workforce planning

    Give short insights.
    """
    return call_llm(prompt)
```

```python
def manager_agent(goal):
    print("Manager: Analyzing goal...")

    # Step 1: Decide which workers to call
    decision_prompt = f"""
    You are a Manager Agent.

    Goal: {goal}

    Decide which workers are needed from:
    [market, finance, operations, legal, hr]

    Return ONLY a Python list.
    Example: ["market", "finance"]
    """

    decision = call_llm(decision_prompt)

    print("Manager decided:", decision)

    results = {}

    # Step 2: Call workers dynamically
    if "market" in decision:
        results["market"] = market_research_worker(goal)

    if "finance" in decision:
        results["finance"] = finance_worker(goal)

    if "operations" in decision:
        results["operations"] = operations_worker(goal)

    if "legal" in decision:
        results["legal"] = legal_worker(goal)

    if "hr" in decision:
        results["hr"] = hr_worker(goal)

    # Step 3: Final decision
    final_prompt = f"""
    You are a Strategy Head.

    Goal: {goal}

    Worker Reports:
    {results}

    Give final decision:
    - Should we launch?
    - Key risks
    - Strategy
    """

    final_output = call_llm(final_prompt)

    return final_output
```

```python
goal = "Evaluate feasibility of launching Product Y in Asia"

result = manager_agent(goal)

print("\nFINAL STRATEGY:\n")
print(result)
```
