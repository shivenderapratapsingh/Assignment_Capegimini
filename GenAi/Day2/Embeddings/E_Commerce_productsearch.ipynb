```python
from sentence_transformers import SentenceTransformer
# Load embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Patient records stored in hospital database
records = [
    "cattle",
    "fridge",
    "washing machine",
    "bulb"
]

# Convert records into embeddings (vectors)
record_vectors = model.encode(records)

# Doctor query
query1 = "light"
query2="clean clothes"

# Convert query into embedding
query_vector1 = model.encode([query1])
query_vector2 = model.encode([query2])

# Calculate similarity between query and records
similarities1 = model.similarity(query_vector1, record_vectors)
similarities2 = model.similarity(query_vector2, record_vectors)

# Print results for query 1
for record, score in zip(records, similarities1[0]):
    print(f"{record} -> Similarity Score: {score:.3f}")
# Print results for query 2
for record, score in zip(records, similarities2[0]):
    print(f"{record} -> Similarity Score: {score:.3f}")

# Find best match
print(f'this is for query 1:{query1}')
best_index = similarities1.argmax()
print("Most Relevant Record:", records[best_index])


# Find best match
print(f'\nthis is for query 2:{query2}')
best_index = similarities2.argmax()
print("Most Relevant Record:", records[best_index])
```
