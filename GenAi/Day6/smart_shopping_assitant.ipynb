```python
# Scenario: Smart Shopping Assistant 🛒
# You say to your students:
# “Imagine you are planning a party and you tell your assistant:
# ‘Find me the best chocolate cake under ₹500.’
# Now instead of just replying with text, your assistant actually thinks, checks different bakery websites, compares prices, and then gives you the final decision.”
```

```python
cakes_db = [
    {"name": "Chocolate Truffle Cake", "price": 499, "rating": 4.6},
    {"name": "Dark Chocolate Cake", "price": 450, "rating": 4.2},
    {"name": "Premium Chocolate Cake", "price": 520, "rating": 4.8},
    {"name": "Budget Chocolate Cake", "price": 300, "rating": 3.9},
]
```

```python
#Agent Brain
def search_cakes(query):
    results = []
    for cake in cakes_db:
        if "chocolate" in query.lower():
            results.append(cake)
    return results
```

```python
def planner(user_query):
    print("🧠 Planning...")

    query = user_query.lower()
    
    # Extract budget
    budget = None
    for word in query.split():
        if word.isdigit():
            budget = int(word)

    return {
        "search_query": "chocolate cake",
        "budget": budget
    }
```

```python
def decide_best(cakes, budget):
    print("📊 Evaluating options...")

    # Filter by budget
    filtered = [cake for cake in cakes if cake["price"] <= budget]

    if not filtered:
        return "❌ No cakes found within budget"

    # Sort by rating
    best = sorted(filtered, key=lambda x: x["rating"], reverse=True)[0]

    return best
```

```python
def shopping_agent(user_query):
    print(f"\n🧾 User Query: {user_query}")

    # Step 1: Plan
    plan = planner(user_query)

    # Step 2: Tool call
    cakes = search_cakes(plan["search_query"])
    print(f"🔍 Found {len(cakes)} cakes")

    # Step 3: Decide
    best = decide_best(cakes, plan["budget"])

    # Step 4: Final Output
    if isinstance(best, dict):
        return f"✅ Best choice: {best['name']} (₹{best['price']}, ⭐{best['rating']})"
    else:
        return best
```

```python
text=input("Enter your quer here?")
result = shopping_agent(text)
print(result)
```
