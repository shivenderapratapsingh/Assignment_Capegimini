```python
# Scenario: Fitness & Wellness Chatbot Training
# You feed it these sentences:
# - “Push-ups strengthen chest and triceps”
# - “Yoga improves flexibility and balance”
# - “Apples are rich in fiber”
# - “Cardio exercises boost heart health”
```

```python
from sentence_transformers import SentenceTransformer
# Load embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
records = ["Push-ups strengthen chest and triceps",
 "Yoga improves flexibility and balance",
 "Apples are rich in fiber",
 "Cardio exercises boost heart health"
]
```

```python
# Convert records into embeddings (vectors)
record_vectors = model.encode(records)
```

```python
# Doctor query
query1 = "push-up is very helpful"
```

```python
query_vector1 = model.encode([query1])
```

```python
similarities1 = model.similarity(query_vector1, record_vectors)
```

```python
for record, score in zip(records, similarities1[0]):
    print(f"{record} -> Similarity Score: {score:.3f}")
```

```python
# Find best match
print(f'this is for query 1:{query1}')
best_index = similarities1.argmax()
print("Most Relevant Record:", records[best_index])
```
