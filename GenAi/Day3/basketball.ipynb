```python
# Scenario: Basketball Player Evaluation
# A coach is evaluating how well a player’s skills match the team’s current needs.
# - The team’s requirement vector is a = [2,5,1].
# - 2 = need for shooting ability
# - 5 = need for defense
# - 1 = need for passing
# - The player’s skill vector is b = [3,4,2].
# - 3 = player’s shooting ability
# - 4 = defensive ability
# - 2 = passing ability
# The coach calculates the dot product:
```

```python
#example of Dot Product (Fastest Similarity Method)
import numpy as np

a = np.array([2,5,1])
b = np.array([3,4,2])

print(np.dot(a,b))
```
