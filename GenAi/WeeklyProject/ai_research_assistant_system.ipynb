```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv

# load API key from .env
load_dotenv()

# initialize model
model = init_chat_model(
    "mistralai/devstral-2-123b-instruct-2512",
    model_provider="nvidia"
)

# sample article
article = """
Artificial Intelligence (AI) is rapidly transforming modern industries including healthcare,
education, and finance. Machine learning algorithms allow computers to learn patterns from
large datasets and make intelligent predictions. In healthcare, AI systems assist doctors in
diagnosing diseases earlier and more accurately. In education, AI helps personalize learning
experiences for students. However, the rise of AI also raises ethical concerns such as data
privacy, algorithm bias, and job displacement caused by automation.
"""

prompt = f"Summarize the following article:\n{article}"

response = model.invoke(prompt)

print(response.content)
```

```python
# Task 2 – Prompt Engineering Experiments
```

```python
# 1 Zero-Shot Prompt
```

```python
prompt = f"""
Summarize the following article in 5 bullet points:

{article}
"""

response = model.invoke(prompt)
print(response.content)
```

```python
# 2 Few-Shot Prompt
prompt = f"""
Example 1
Article: Electric vehicles reduce pollution and reliance on fossil fuels.

Summary:
• EVs reduce carbon emissions
• Promote sustainable transportation
• Support renewable energy use

Example 2
Article: Climate change increases global temperatures and causes extreme weather.

Summary:
• Rising global temperatures
• Increased natural disasters
• Need for global climate policies

Now summarize the following article:

{article}
"""

response = model.invoke(prompt)
print(response.content)
```

```python
#3. Chain-of-Thought Prompt
prompt = f"""
Analyze the article step by step:

1. Identify the main topic
2. Extract key ideas
3. Generate a concise summary

Article:
{article}
"""

response = model.invoke(prompt)
print(response.content)
```

```python
# Task 3 – Prompt Optimization
optimized_prompt = f"""
Read the article and generate a professional executive summary.

Your response must include:

1. Three key insights
2. One actionable recommendation
3. Professional tone

Article:
{article}
"""

response = model.invoke(optimized_prompt)
print(response.content)
```

```python
# Task 4 – Tokenization Experiment
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "Artificial Intelligence helps computers learn from data."

tokens = tokenizer.tokenize(text)

print("Tokens:", tokens)
print("Token Count:", len(tokens))
```

```python
# Task 5 – Build a Mini AI Tool


model = init_chat_model(
    "mistralai/devstral-2-123b-instruct-2512",
    model_provider="nvidia"
)

def research_assistant(article_text):

    prompt = f"""
    Analyze the article and provide:

    1. Short Summary
    2. Key Insights
    3. One Actionable Recommendation

    Article:
    {article_text}
    """

    response = model.invoke(prompt)

    return response.content


# user input
article = input("Enter article text:\n")

result = research_assistant(article)


print("\nAI Research Assistant Output:\n")
print(result)
```
